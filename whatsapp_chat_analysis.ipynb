```python
f = open("WhatsApp Chat with CSE-2 unoff.txt", encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s?(?:am|pm)\s-\s'
```

```python
messages = re.split(pattern , data)[1:]
```

```python
messages
```

```python
dates = re.findall(pattern,data)
```

```python
dates
```

```python
print(len(dates) , len(messages))
```

```python
df = pd.DataFrame({
    'user_message': messages,
    'message_date': dates
})

df['message_date'] = pd.to_datetime(
    df['message_date']
        .str.replace('\u202f', ' ', regex=False)
        .str.strip(),
    format='%d/%m/%y, %I:%M %p -'
)


df.rename(columns={'message_date': 'date'}, inplace=True)

df.head()
```

```python
df.sample(10)
```

```python
# seperating users and messages

users = []
messages = []

for i in df['user_message']:
    entry = re.split(
        r'((?:\+?\d[\d\s-]{7,})|(?:[A-Za-z][A-Za-z\s]*)):\s',
        i,
        maxsplit=1
    )

    ## append users and messages in their respective coln 
    if len(entry) > 1:
        users.append(entry[1])
        messages.append(entry[2])
    ## if it's not a user and a message is their it is a group notification
    else:
        users.append("group_notification")
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
```

```python
period = []

for hour in df[['day_name','hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour==00:
        period.append(str('00') + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
```

```python
df['period'] = period
```

```python
df.head()
```

```python
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday',
             'Friday', 'Saturday', 'Sunday']

period_order = sorted(df['period'].unique(),
                      key=lambda x: int(x.split('-')[0]))

```

```python
import seaborn as sns

pivot = df.pivot_table(
    index='day_name',
    columns='period',
    values='message',
    aggfunc='count'
).fillna(0)

pivot = pivot.reindex(day_order).reindex(columns=period_order)

plt.figure(figsize=(20,6))
sns.heatmap(pivot)
plt.show()
```
